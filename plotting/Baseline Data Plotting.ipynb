```python
%matplotlib notebook

import matplotlib.pyplot as plt
from IPython.display import display
import neurokit2 as nk
import pandas as pd
import numpy as np
import seaborn as sns
import os
import pickle

# Set matplotlib parameters for displaying graphs
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = [16,8]  # Bigger images
plt.rcParams['font.size']= 14

pd.set_option('display.max_columns', None)

part_number = '201'
sampling_rate=2000

save_dir = '../data/part'+part_number+'/figures/'

baseline_fn = '../data/part'+part_number+'/part'+part_number+'_baseline_ecg_signals.p'
baseline2_fn = '../data/part'+part_number+'/part'+part_number+'_baseline_eda_signals.p'
baseline3_fn = '../data/part'+part_number+'/part'+part_number+'_baseline_rsp_signals.p'
ecg_baseline = pickle.load(open(baseline_fn,"rb"))
eda_baseline = pickle.load(open(baseline2_fn,"rb"))
rsp_baseline = pickle.load(open(baseline3_fn,"rb"))

x_axis = np.linspace(0, ecg_baseline.shape[0]/ sampling_rate, ecg_baseline.shape[0])
```

```python
fig = nk.ecg_plot(ecg_baseline, sampling_rate=sampling_rate)
plt.savefig(save_dir+part_number+ ' ECG Baseline.jpg')
```

```python
plt.figure()
plt.plot(x_axis,eda_baseline['EDA_Clean'])
plt.savefig(save_dir+part_number+ ' EDA Baseline.jpg')
```

```python
fig = nk.rsp_plot(rsp_baseline, sampling_rate=sampling_rate)
plt.savefig(save_dir+part_number+ ' RSP Baseline.jpg')
```
